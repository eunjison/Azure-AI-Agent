# 06. Simple Evaluation  (수정필요)

간단한 **정확도 유사 지표** 예시를 통해 모델 출력 비교 방식을 익힙니다.


```python
def simple_overlap_score(reference: str, hypothesis: str) -> float:
    ref = set(reference.lower().split())
    hyp = set(hypothesis.lower().split())
    return 0.0 if not ref else len(ref & hyp) / len(ref)
```

```python
samples = [
    {"ref": "오늘 날씨는 맑고 기온은 25도입니다.", "hyp": "오늘 맑고 기온은 25도야."},
    {"ref": "안녕하세요, 무엇을 도와드릴까요?", "hyp": "안녕하세요. 어떤 도움이 필요하신가요?"},
    {"ref": "저녁 7시에 운동을 갑니다.", "hyp": "저녁 일곱 시에 운동할 예정입니다."},
]
for s in samples:
    print("Ref:", s["ref"])
    print("Hyp:", s["hyp"])
    print("Score:", round(simple_overlap_score(s["ref"], s["hyp"]), 3))
    print("-"*30)
```
