```python
def __init__(self, client: ChatCompletionClient, hil_mode: bool = False):
    self.client = client
    self._validate_client_capabilities(client)

    fs = FileSurfer("FileSurfer", model_client=client)
    ws = MultimodalWebSurfer("WebSurfer", model_client=client)
    coder = MagenticOneCoderAgent("Coder", model_client=client)
    executor = CodeExecutorAgent("Executor", code_executor=LocalCommandLineCodeExecutor())
    agents: List[ChatAgent] = [fs, ws, coder, executor]
    if hil_mode:
        user_proxy = UserProxyAgent("User")
        agents.append(user_proxy)
    super().__init__(agents, model_client=client)

```

````python
# üìä Í∞ÑÏÜåÌôîÎêú Î©ÄÌã∞ ÏóêÏù¥Ï†ÑÌä∏ ÏãúÏä§ÌÖú (Docker ÏóÜÏù¥)
import yfinance as yf
import requests
import json
from datetime import datetime
import matplotlib.pyplot as plt

# Ìè∞Ìä∏ ÏÑ§Ï†ï (ÌïúÍ∏Ä Í≤ΩÍ≥† Î∞©ÏßÄ)
plt.rcParams['font.family'] = 'DejaVu Sans'
plt.rcParams['axes.unicode_minus'] = False

class SimpleWebSurfer:
    """Í∞ÑÎã®Ìïú Ïõπ ÏÑúÌïë ÏóêÏù¥Ï†ÑÌä∏"""
    def __init__(self):
        self.name = "WebSurfer"
    
    def get_stock_data(self, symbol="MSFT"):
        """Ï£ºÏãù Îç∞Ïù¥ÌÑ∞ ÏàòÏßë"""
        try:
            stock = yf.Ticker(symbol)
            data = stock.history(period="1y")
            info = stock.info
            
            current_price = data['Close'].iloc[-1]
            year_high = data['High'].max()
            year_low = data['Low'].min()
            annual_return = ((current_price / data['Close'].iloc[0]) - 1) * 100
            
            return {
                "symbol": symbol,
                "current_price": current_price,
                "year_high": year_high,
                "year_low": year_low,
                "annual_return": annual_return,
                "market_cap": info.get('marketCap', 'N/A'),
                "pe_ratio": info.get('trailingPE', 'N/A'),
                "data": data
            }
        except Exception as e:
            return {"error": f"Îç∞Ïù¥ÌÑ∞ ÏàòÏßë Ïã§Ìå®: {e}"}

class SimpleAnalyst:
    """Í∞ÑÎã®Ìïú Î∂ÑÏÑù ÏóêÏù¥Ï†ÑÌä∏"""
    def __init__(self):
        self.name = "Analyst"
    
    def analyze_stock(self, stock_data):
        """Ï£ºÏãù Î∂ÑÏÑù"""
        if "error" in stock_data:
            return stock_data
        
        analysis = {
            "ÌòÑÏû¨_ÏÉÅÌô©": "Î∂ÑÏÑù ÏßÑÌñâ Ï§ë...",
            "Í∏∞Ïà†Ï†Å_Î∂ÑÏÑù": {},
            "Ìà¨Ïûê_ÏùòÍ≤¨": "",
            "Î™©ÌëúÍ∞Ä": 0
        }
        
        # Í∏∞Ïà†Ï†Å Î∂ÑÏÑù
        data = stock_data["data"]
        recent_20d = data['Close'].tail(20).mean()
        recent_50d = data['Close'].tail(50).mean()
        
        analysis["Í∏∞Ïà†Ï†Å_Î∂ÑÏÑù"] = {
            "20Ïùº_ÌèâÍ∑†": recent_20d,
            "50Ïùº_ÌèâÍ∑†": recent_50d,
            "Ï∂îÏÑ∏": "ÏÉÅÏäπ" if recent_20d > recent_50d else "ÌïòÎùΩ",
            "ÌòÑÏû¨Í∞Ä_vs_20ÏùºÌèâÍ∑†": ((stock_data["current_price"] / recent_20d) - 1) * 100
        }
        
        # Ìà¨Ïûê ÏùòÍ≤¨
        if analysis["Í∏∞Ïà†Ï†Å_Î∂ÑÏÑù"]["Ï∂îÏÑ∏"] == "ÏÉÅÏäπ":
            analysis["Ìà¨Ïûê_ÏùòÍ≤¨"] = "Í∏çÏ†ïÏ†Å - ÏÉÅÏäπ Ï∂îÏÑ∏ ÏßÄÏÜç"
            analysis["Î™©ÌëúÍ∞Ä"] = stock_data["current_price"] * 1.1
        else:
            analysis["Ìà¨Ïûê_ÏùòÍ≤¨"] = "Ïã†Ï§ëÌï® - ÌïòÎùΩ Ï∂îÏÑ∏ ÌôïÏù∏ ÌïÑÏöî"
            analysis["Î™©ÌëúÍ∞Ä"] = stock_data["current_price"] * 0.95
        
        return analysis

class SimpleVisualizer:
    """Í∞ÑÎã®Ìïú ÏãúÍ∞ÅÌôî ÏóêÏù¥Ï†ÑÌä∏"""
    def __init__(self):
        self.name = "Visualizer"
    
    def create_chart(self, stock_data):
        """Ï∞®Ìä∏ ÏÉùÏÑ±"""
        if "error" in stock_data:
            print(f"‚ùå Ï∞®Ìä∏ ÏÉùÏÑ± Ïã§Ìå®: {stock_data['error']}")
            return
        
        data = stock_data["data"]
        
        plt.figure(figsize=(12, 8))
        
        # Ï£ºÍ∞Ä Ï∞®Ìä∏
        plt.subplot(2, 1, 1)
        plt.plot(data.index, data['Close'], linewidth=2, color='#1f77b4')
        plt.title(f'{stock_data["symbol"]} Stock Price (1 Year)', fontsize=14, fontweight='bold')
        plt.ylabel('Price ($)')
        plt.grid(True, alpha=0.3)
        
        # ÌòÑÏû¨Í∞Ä ÌëúÏãú
        plt.axhline(y=stock_data["current_price"], color='red', linestyle='--', alpha=0.7)
        plt.text(data.index[-1], stock_data["current_price"], 
                f'Current: ${stock_data["current_price"]:.2f}',
                verticalalignment='bottom', fontweight='bold')
        
        # Í±∞ÎûòÎüâ Ï∞®Ìä∏
        plt.subplot(2, 1, 2)
        plt.plot(data.index, data['Volume'], color='orange', alpha=0.7)
        plt.title('Trading Volume', fontsize=12, fontweight='bold')
        plt.ylabel('Volume')
        plt.grid(True, alpha=0.3)
        
        plt.tight_layout()
        plt.show()

# üöÄ Í∞ÑÏÜåÌôîÎêú MagenticOne ÌåÄ Ïã§Ìñâ
print("ü§ñ Í∞ÑÏÜåÌôîÎêú Î©ÄÌã∞ ÏóêÏù¥Ï†ÑÌä∏ ÌåÄ Ï¥àÍ∏∞Ìôî...")

# ÏóêÏù¥Ï†ÑÌä∏ ÏÉùÏÑ±
web_surfer = SimpleWebSurfer()
analyst = SimpleAnalyst()
visualizer = SimpleVisualizer()

print(f"‚úÖ {web_surfer.name} ÏóêÏù¥Ï†ÑÌä∏ Ï§ÄÎπÑ ÏôÑÎ£å")
print(f"‚úÖ {analyst.name} ÏóêÏù¥Ï†ÑÌä∏ Ï§ÄÎπÑ ÏôÑÎ£å") 
print(f"‚úÖ {visualizer.name} ÏóêÏù¥Ï†ÑÌä∏ Ï§ÄÎπÑ ÏôÑÎ£å")

# ÏûëÏóÖ ÏàòÌñâ
print("\nüìã ÏûëÏóÖ: ÎßàÏù¥ÌÅ¨Î°úÏÜåÌîÑÌä∏ ÌòÑÏû¨ Ï£ºÍ∞ÄÏôÄ Ï†ÑÎßù Î∂ÑÏÑù")
print("üîç 1Îã®Í≥Ñ: Îç∞Ïù¥ÌÑ∞ ÏàòÏßë Ï§ë...")

# WebSurferÍ∞Ä Îç∞Ïù¥ÌÑ∞ ÏàòÏßë
msft_data = web_surfer.get_stock_data("MSFT")

if "error" not in msft_data:
    print("‚úÖ Îç∞Ïù¥ÌÑ∞ ÏàòÏßë ÏôÑÎ£å")
    
    print("\nüìä 2Îã®Í≥Ñ: Î∂ÑÏÑù ÏßÑÌñâ Ï§ë...")
    # AnalystÍ∞Ä Î∂ÑÏÑù ÏàòÌñâ
    analysis = analyst.analyze_stock(msft_data)
    
    print("‚úÖ Î∂ÑÏÑù ÏôÑÎ£å")
    
    print("\nüìà 3Îã®Í≥Ñ: ÏãúÍ∞ÅÌôî ÏÉùÏÑ± Ï§ë...")
    # VisualizerÍ∞Ä Ï∞®Ìä∏ ÏÉùÏÑ±
    visualizer.create_chart(msft_data)
    
    # Í≤∞Í≥º Ï∂úÎ†•
    print("\n" + "="*60)
    print("üìä MICROSOFT (MSFT) Ï¢ÖÌï© Î∂ÑÏÑù Í≤∞Í≥º")
    print("="*60)
    
    print(f"\nüí∞ Ï£ºÍ∞Ä Ï†ïÎ≥¥:")
    print(f"   ÌòÑÏû¨Í∞Ä: ${msft_data['current_price']:.2f}")
    print(f"   52Ï£º ÏµúÍ≥†: ${msft_data['year_high']:.2f}")
    print(f"   52Ï£º ÏµúÏ†Ä: ${msft_data['year_low']:.2f}")
    print(f"   Ïó∞Í∞Ñ ÏàòÏùµÎ•†: {msft_data['annual_return']:+.1f}%")
    
    if isinstance(msft_data['market_cap'], (int, float)):
        market_cap_b = msft_data['market_cap'] / 1e9
        print(f"   ÏãúÍ∞ÄÏ¥ùÏï°: ${market_cap_b:.1f}B")
    
    if isinstance(msft_data['pe_ratio'], (int, float)):
        print(f"   PER: {msft_data['pe_ratio']:.1f}")
    
    print(f"\nüîç Í∏∞Ïà†Ï†Å Î∂ÑÏÑù:")
    tech_analysis = analysis["Í∏∞Ïà†Ï†Å_Î∂ÑÏÑù"]
    print(f"   20Ïùº ÌèâÍ∑†: ${tech_analysis['20Ïùº_ÌèâÍ∑†']:.2f}")
    print(f"   50Ïùº ÌèâÍ∑†: ${tech_analysis['50Ïùº_ÌèâÍ∑†']:.2f}")
    print(f"   Ï∂îÏÑ∏: {tech_analysis['Ï∂îÏÑ∏']}")
    print(f"   20Ïùº ÌèâÍ∑† ÎåÄÎπÑ: {tech_analysis['ÌòÑÏû¨Í∞Ä_vs_20ÏùºÌèâÍ∑†']:+.1f}%")
    
    print(f"\nüí° Ìà¨Ïûê ÏùòÍ≤¨:")
    print(f"   ÏùòÍ≤¨: {analysis['Ìà¨Ïûê_ÏùòÍ≤¨']}")
    print(f"   Î™©ÌëúÍ∞Ä: ${analysis['Î™©ÌëúÍ∞Ä']:.2f}")
    
    print(f"\nüéØ ÏöîÏïΩ:")
    if analysis["Í∏∞Ïà†Ï†Å_Î∂ÑÏÑù"]["Ï∂îÏÑ∏"] == "ÏÉÅÏäπ":
        summary = "MicrosoftÎäî ÌòÑÏû¨ ÏÉÅÏäπ Ï∂îÏÑ∏Î•º Î≥¥Ïù¥Í≥† ÏûàÏúºÎ©∞, Í∏∞Ïà†Ï†Å ÏßÄÌëúÍ∞Ä Í∏çÏ†ïÏ†ÅÏûÖÎãàÎã§."
    else:
        summary = "MicrosoftÎäî Îã®Í∏∞Ï†ÅÏúºÎ°ú Ï°∞Ï†ï Íµ≠Î©¥Ïóê ÏûàÏúºÎãà Ïã†Ï§ëÌïú Ï†ëÍ∑ºÏù¥ ÌïÑÏöîÌï©ÎãàÎã§."
    
    print(f"   {summary}")
    
    print(f"\n‚úÖ Î©ÄÌã∞ ÏóêÏù¥Ï†ÑÌä∏ Î∂ÑÏÑù ÏôÑÎ£å!")

else:
    print(f"‚ùå ÏûëÏóÖ Ïã§Ìå®: {msft_data['error']}")

print("\nüí° Ïù¥ ÏòàÏ†úÎäî Docker ÏóÜÏù¥ÎèÑ Î©ÄÌã∞ ÏóêÏù¥Ï†ÑÌä∏ ÌòëÏóÖÏùÑ Íµ¨ÌòÑÌïú Í∞ÑÏÜåÌôî Î≤ÑÏ†ÑÏûÖÎãàÎã§!")
print("üöÄ Ïã§Ï†ú MagenticOne ÏÇ¨Ïö©ÏùÑ ÏõêÌïòÏãúÎ©¥ Docker Ìå®ÌÇ§ÏßÄÎ•º ÏÑ§ÏπòÌï¥Ï£ºÏÑ∏Ïöî.")
````
